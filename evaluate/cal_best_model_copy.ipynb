## 排序模型的表現，這個程式使用 F1@3-Alike@3 最高的模型當作表現最好的模型，需要讀取的檔案為 results 內算好分數的檔案

```python
import json
from pprint import pprint
def open_jsonl_score(name):
    with open('{}.jsonl'.format(name),'r',encoding='UTF-8') as f:
        data = [json.loads(line) for line in f]
    return data
dataset = open_jsonl_score("model_score")
```

## 計算 F1@3 - alike@3，並且使用 F1@3 做排序，因為我們想要在不損失 DG 表現下最好的模型

```python
len_dataset = len(dataset)
for i in range(len_dataset):
    dataset[i]["f1-a-alike"] = dataset[i]["F1@3"] - dataset[i]["alike@3"]
dataset_sorted = sorted(dataset,key = lambda e:e["F1@3"],reverse=True)
```
